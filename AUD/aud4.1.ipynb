```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno
from sklearn.impute import SimpleImputer, KNNImputer
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
```

```python
data = pd.read_csv('spaceship_titanic.csv')
data1 = data.copy()
data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
missing_percent = data.isnull().sum() / len(data) * 100
missing_percent
```

```python
data = data.dropna(subset=['RoomService'], axis=0)
data.isnull().sum()
```

```python
data = data.drop(columns='Cabin')
data.isnull().sum()
```

```python
constant = SimpleImputer(strategy='constant',fill_value=0)
data['FoodCourt'] = constant.fit_transform([data['FoodCourt']])[0]
data.isnull().sum()
```

```python
frequent = SimpleImputer(strategy='most_frequent')
data['HomePlanet'].value_counts().plot.bar()
plt.show()
```

```python
data['HomePlanet'] = data['HomePlanet'].astype(str)
data['HomePlanet'] = frequent.fit_transform(data['HomePlanet'].values.reshape(1,-1))[0]
data.isnull().sum()
```

```python
data.Age.plot.hist()
plt.show()
```

```python
data['Age'] = data['Age'].fillna(data.Age.mean())
```

```python
data.isnull().sum()
```

```python
data.ShoppingMall.plot.hist()
```

```python
data['ShoppingMall'] = data['ShoppingMall'].fillna(data.ShoppingMall.median())
data.isnull().sum()
```

```python
#KNN
```

```python
data[['VRDeck','Spa']].plot.hist(subplots=True)
plt.show()
```

```python
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(data[['VRDeck','Spa']])
knn = KNNImputer(n_neighbors=5)
imputed_data = knn.fit_transform(scaled_data)
imputed_data
```

```python
imputed_data = scaler.inverse_transform(imputed_data)
imputed_data
```

```python
data[['VRDeck','Spa']] = imputed_data
data.isnull().sum()
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
```

```python
mice = IterativeImputer()
imputed_values = mice.fit_transform(data[['VRDeck', 'FoodCourt', 'Spa']])
data[['VRDeck', 'FoodCourt', 'Spa']] = imputed_values
```
